'Bartok', 'Bellini', 'Buxtehude', 'Bernstein'. Which one comes last?

class Program
    {
        static void Main(string[] args)
        {
            Console.Write("Please enter the words seperated by commas and without any space or special charaters : ");
            string WordsEntered = Console.ReadLine();
            string[] Words = WordsEntered.Split(',');
            var EnteredWords = new SortedSet<string>();

            for (int i = 0; i < Words.Length; i++)
            {
                EnteredWords.Add(Words[i]);
            }

            string[] SortedWords = new string[Words.Length];
            int temp = 0;
            foreach (var word in EnteredWords)
            {
                if (temp < Words.Length)
                {
                    SortedWords[temp] = word;
                }
                temp++;

            }
            Console.WriteLine("In the dictionary, {0} comes first and {1} comes last",
                SortedWords[0], SortedWords[SortedWords.Length - 1]);
            Console.ReadLine();
        }

    }

6. You can turn a word into pig-Latin using the following two rules (simplified):
 If the word starts with a consonant, move that letter to the end and append
'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
 If the word starts with a vowel, simply append 'way' to the end of the word.
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case
characters. Your output should always be lower case however.

>>> pig('happy')
'appyhay'
>>> pig('Enter')
'enterway' 

```python
def pig(name):
    if (name[0]=='A'or name[0]=='E'or name[0]=='I'or name[0]=='O'or name[0]=='U'
        or name[0]=='a'or name[0]=='e'or name[0]=='i'or name[0]=='o'or name[0]=='u'):
        word= name+'way'
        return (word.lower())

    else:
        word= name[1:]+name[0]+'ay'
        return (word.lower())


name= input("Please enter the word: ")
pig(name)
```

File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.
Write a function bldcount() that reads the file with name name and reports (i.e.,
prints) how many patients there are in each bloodtype.
>>> bldcount('bloodtype.txt')
There are 10 patients of blood type A.
There is one patient of blood type B.
There are 10 patients of blood type AB.
There are 12 patients of blood type O.
There are no patients of blood type OO.

```python
def bldcount(path):
    infile = open (path)
    content=infile.read()
    infile.close()
    wordlist=content.split()
    blooda = wordlist.count("A")
    bloodb = wordlist.count("B")
    bloodab = wordlist.count("AB")
    bloodo = wordlist.count("O")
    bloodoo = wordlist.count("OO")
    print ("There are {} patients of blood type A".format(blooda))
    print ("There are {} patients of blood type B".format(bloodb))
    print ("There are {} patients of blood type AB".format(bloodab))
    print ("There are {} patients of blood type O".format(bloodo))
    print ("There are {} patients of blood type OO".format(bloodoo))

path= input ("Please enter the full path of the file: ")
bldcount(path)
```
